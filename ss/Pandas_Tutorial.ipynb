```python
import datetime

def round_time(dt, i_min):
    a, b = divmod(dt.minute, i_min)
    c = i_min * (a + round(b / i_min, 0))
    hours, minutes= divmod(c, 60)
    hours = datetime.timedelta(hours = hours)
    minutes = datetime.timedelta(minutes = minutes)
    return datetime.datetime(dt.year, dt.month, dt.day, dt.hour, 0, 0) + hours + minutes

raw_time = '2020/12/01 23:49:00'
time = datetime.datetime.strptime(raw_time, '%Y/%m/%d %H:%M:%S')
time = round_time(time, 20)
print(time)
```

```python
poketype_secondary = df['Type 1'].value_counts()
poketype_secondary
```

```python
flyer_df = pd.concat([poketype_secondary_flyer, poketype_secondary], axis='columns', sort=False)
flyer_df.rename(columns={'Type 2': "No_of_Flyers", 'Type 1': 'Total_of_Type'}, inplace=True)
flyer_df['Percentage_Flyers'] = (flyer_df['No_of_Flyers'] / flyer_df['Total_of_Type'])*100
flyer_df.sort_values(by='Percentage_Flyers', ascending=False, inplace=True)
flyer_df
```
